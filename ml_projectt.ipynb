```python
# Gereken kütüphaneler eklendi.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Datanın çıktısını almak için 'head' kullanılır.Var olan standart değer ilk 5 kayıttır. 

df.head()
```

```python
# sns.countplot komutu; sütun grafiği ile kategorik bir değişkenin her değerine sahip veri noktalarının sayısını gösterir.
 
sns.countplot(x='left' ,data=df)
plt.show()
```

```python
# Maaş durumu kategorilendirildi. 

sns.countplot(x='salary' ,data=df)
plt.show()
```

```python
# Düşük maaşlı kişilerin işten ayrılma oranının yüksek olduğu görünüyor. 

sns.catplot(x='left', col='salary', kind= 'count', data=df); 
```

```python
# Son 5 yıl da promosyon alan çalışan sayısının yüksek olduğu görünüyor.
sns.countplot(x='promotion_last_5years' ,data=df)
plt.show() 
```

```python
# İşten ayrılma durumuna göre sütunlar bölündü ve terfi etmeyen insanlar daha fazla ayrıldı. 
sns.catplot(x='left', col='promotion_last_5years', kind='count' , data=df);
```

```python
# İş kazası yapan insanların ayrılması gösterildi.

sns.countplot(x='Work_accident', data=df)
```

```python
sns.catplot(x='left', col='Work_accident', kind='count' , data=df);
```

```python
sns.displot(df['satisfaction_level'])
```

```python
# Memmnuniyet seviyesindeki gözlemler tek değişkenli dağılımda çizildi.
# 0 - 1 arasında olan memnuniyetlik duruma bakıldığında seviye olarak 0.1 de yüksek sayıya sahiptir.

sns.displot(df['satisfaction_level'])
```

```python
#Kategorilere göre dağılımı gösteren kutu grafiği çizildi.
#0.2 ile 0.5 arasında az memnuniyette olan çalışanlar en çok işi bırakanlar olmuş.  

sns.boxplot(x='left', y='satisfaction_level' , data=df)
```

```python
#Proje sayısının 2 den 7 ye dağılımı yapıldı. En çok proje sayısı seviye 4 tedir. 

sns.displot(df['number_project'])
```

```python
# İş yerinde geçirilen ortalama aylık saati yüksek olan çalışanlar ayrılmıştır.

sns.boxplot(x='left', y='average_montly_hours' , data=df) 
```

```python
sns.displot(df['time_spend_company'])
```

```python
# Korelasyon kontrolü yapıldı. Matris tarzında veriler görselleştirildi.

cor=df.corr()
fig=plt.figure(figsize=(15,7))
sns.heatmap(cor,annot=True)
```

```python
mask = np.zeros_like(df.corr())
mask[np.tril_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, vmax=1, square=True,annot=True,cmap='viridis', mask=mask.T)

plt.title('Correlation between different fearures')
```
